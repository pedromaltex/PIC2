```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import help
from yahooquery import Screener
```

```python
def get_companiesbysector(sector): # Obtém 100 maiores empresas do determinado setor
    sector_indices = [
    'ms_basic_materials',
    'ms_communication_services',
    'ms_consumer_cyclical',
    'ms_consumer_defensive',
    'ms_energy',
    'ms_financial_services',
    'ms_healthcare',
    'ms_industrials',
    'ms_real_estate',
    'ms_technology',
    'ms_utilities',
    'most_visited_basic_materials',
    'most_visited_communication_services',
    'most_visited_consumer_cyclical',
    'most_visited_consumer_defensive',
    'most_visited_energy',
    'most_visited_financial_services',
    'most_visited_healthcare',
    'most_visited_industrials',
    'most_visited_real_estate',
    'most_visited_technology',
    'most_visited_utilities'
]

        
    # Obtém as empresas do setor escolhido
    screener = Screener()
    data = screener.get_screeners(sector, count=100)  # Pode ajustar 'count'

    #print(data[sector]['quotes'][0]['symbol'])

    # Extraindo os símbolos das empresas
    sector_tickers = [stock['symbol'] for stock in data[sector]['quotes']]
    return sector_tickers
```

```python
ticker1 = get_companiesbysector('ms_energy')
ticker2 = get_companiesbysector('ms_financial_services')
ticker2
```

```python
# Baixar os dados históricos para as empresas do S&P 500 + GLD
day, month, year = datetime.now().day, datetime.now().month, datetime.now().year
start_date = f"{year-99}-{month}-{day}"  # Começo de 99 anos atrás
end_date = f"{year}-{month}-{day}"

# Baixar os dados
dataset = help.dowl_data_return_dataset(ticker1, ticker2, start_date, end_date)
```

```python
# Preencher dados ausentes com zeros
dataset = dataset.fillna(1)
dataset
```

```python
# Calcular as percentagens diárias
percentagens_dataset = help.calc_returns_daily(dataset)
percentagens_dataset
```

```python
# Calcular as correlações anuais
correlation_by_year = {}
for years_ago in range(31):
    # Filtrar os dados para o ano específico
    new_data = percentagens_dataset[percentagens_dataset.index.year == year - years_ago]
    
    # Garantir que todas as ações têm os mesmos índices de datas
    new_data = new_data.reindex(pd.date_range(start=start_date, end=end_date, freq='B'), method='bfill')  # ou 'bfill'
    
    # Calcular a correlação entre o ouro (GLD) e todas as empresas do S&P 500
    correlations = {}
    for ticker in ticker1:
        corr = new_data[ticker].corr(new_data[ticker2])  # Correlação direta
        correlations[ticker] = corr
    
    correlation_by_year[f'{year - years_ago}'] = correlations
```

```python
# Transformar as correlações em um DataFrame para visualização
correlation_df = pd.DataFrame(correlation_by_year)
```

```python
# Gerar o heatmap
plt.figure(figsize=(12, 20))
sns.heatmap(correlation_df, annot=False, cmap='coolwarm', center=0, fmt=".2f", linewidths=0, cbar=True)

# Personalizar o gráfico
plt.title('Correlação Anual: Ouro (GLD) vs Empresas do S&P 500')
plt.xlabel('Ano')
plt.ylabel('Empresas do S&P 500')

# Exibir o gráfico
plt.show()
```
